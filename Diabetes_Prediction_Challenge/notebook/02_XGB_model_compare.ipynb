# 🚀 **Diabetes Prediction — XGBoost (Extreme Gradient Boosting)**
Implementação do modelo XGBoost utilizando a API `XGBClassifier`. Este modelo faz parte da estratégia de *Gradient Boosting* escalável, utilizando o método de construção de árvores `hist` para otimização de performance em grandes volumes de dados (210.000 amostras).

**Destaques do Pipeline:**
* **Evolução de Preprocessamento:** Utilização do `PP3` (v3.0), focado em engenharia de atributos avançada e tratamento de variáveis categóricas nativas do XGBoost.
* **Otimização:** Comparação entre Baseline, Modelo 1 (Random Search) e Modelo 2 (Refined Tuning).
* **Rigor Estatístico:** Teste-T pareado realizado para validar se o ganho de performance justifica a complexidade dos novos hiperparâmetros.

## 2. Dataload & Preprocessamento com joblib


```python
BASE = Path.cwd().parent   
# =====================================================
# ⚙️ 0. carregamento dos preprocessador 
# =====================================================
PP3 = joblib.load(BASE/'src'/'preprocess_diabetes_v3.joblib')['preprocessador']

# # =====================================================
# # 📁 1. Leitura dos dados & Separação das bases
# # =====================================================

DATA_DIR = BASE / "data" / "raw"
X_train = pd.read_csv(DATA_DIR / "X_train_raw.csv").reset_index(drop=True)
X_val  = pd.read_csv(DATA_DIR / "X_test_raw.csv")
y_train = pd.read_csv(DATA_DIR / "y_train_raw.csv").values.ravel()
y_val  = pd.read_csv(DATA_DIR / "y_test_raw.csv")
mtd_scoring='roc_auc'

# # # =====================================================
# # #  🤖 3.Definição dos Modelos
# # # =====================================================
model_XGB0 = XGBClassifier(objective='binary:logistic',eval_metric='logloss',enable_categorical=True,
                           tree_method="hist" , random_state=42)
pipe_XGB0      = pipe_models(model_XGB0,PP3)


DATA_MODELS= BASE /"models"
pipe_XGB1 = joblib.load(DATA_MODELS / 'modelo_XGB2_final_randsearch.roc_auc_v3.joblib')
pipe_XGB2 = joblib.load(DATA_MODELS / 'modelo_XGB2_final_refine.roc_auc_v3.joblib')
print("\n#Processo iniciado em:", time.strftime("%H:%M:%S"))
```

## 4 Relatorio de Classificação

```python
# =====================================================
# A. Avaliação comparativa dos modelos RF (Baseline)
# =====================================================
# Nesta etapa é realizada uma análise comparativa completa entre os
# modelos testados. O objetivo é avaliar
# desempenho, estabilidade e diferença estatística entre os modelos.
# Calcula os scores de validação cruzada para cada modelo(roc-auc)

# -----------------------------------------------------
# A.1 Validação cruzada ROC-AUC & acuraccy
# -----------------------------------------------------

s0_auc = R0['cv_scores'][0]
s1_auc = R1['cv_scores'][0]
s2_auc = R2['cv_scores'][0]

cv_s = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)


print(f"Iniciando Cross-Validation Accuracy ...")
start_m = time.time()
s0_acc = cross_val_score(pipe_XGB0, X_train, y_train,
                                 cv=cv_s,
                                 scoring='accuracy',
                                 n_jobs=-1)
print(f"    ✅ Baseline - ⏱️ {time.time() - start_m:5.2f}s")
start_m = time.time()
s1_acc = cross_val_score(pipe_XGB1, X_train, y_train,
                                 cv=cv_s,
                                 scoring='accuracy',
                                 n_jobs=-1)
print(f"    ✅ modelo 1 - ⏱️ {time.time() - start_m:5.2f}s")
start_m = time.time()
s2_acc = cross_val_score(pipe_XGB2, X_train, y_train,
                                 cv=cv_s,
                                 scoring='accuracy',
                                 n_jobs=-1)
print(f"    ✅ modelo 2 - ⏱️ {time.time() - start_m:5.2f}s")

# -----------------------------------------------------
# A.2 Probabilidades no conjunto de teste
# -----------------------------------------------------
y_prob0 = R0['y_probs'][0]
y_prob1 = R1['y_probs'][0]
y_prob2 = R2['y_probs'][0]

# -----------------------------------------------------
# A.4 Estrutura dos dados para relatorio
# -----------------------------------------------------
models_list = [
    ('Modelo 0  (base)' , pipe_XGB0, s0_auc, s0_acc, y_prob0,R0['best_t'][0]),
    ('Modelo 1  (base)' , pipe_XGB0, s0_auc, s0_acc, y_prob0, R0['best_t'][0]),
    ('Modelo 2  (exp) ', pipe_XGB1, s1_auc, s1_acc, y_prob1   ,R1['best_t'][0]),
    ('Modelo 3  (ref) ', pipe_XGB2, s2_auc, s2_acc, y_prob2   , R2['best_t'][0])
]
# -----------------------------------------------------
# A.5 Geração do relatório estatístico
# -----------------------------------------------------

print(f" Construindo relatorio ...")
df_results,W = gerar_relatorio_estatistico2(models_list,X_train, y_train,X_val, y_val)
```

### 5. Considerações Finais

O XGBoost consolidou-se como o modelo de performance superior nesta etapa, extraindo o máximo potencial da engenharia de atributos avançada (v3.0). A análise estatística confirma que o processo de tuning (Modelo 2) superou o Baseline de forma inequívoca, estabelecendo um novo patamar de ROC-AUC para o projeto.

**Análise de Generalização e Significância**
O XGBoost demonstrou uma aderência excepcional entre os ambientes de treino e teste, com uma discrepância quase nula ($\Delta = 0.0001$) entre a média da validação cruzada e o conjunto de teste final. O teste-t pareado validou que o ganho de performance do **Modelo 2 (exp)** sobre o Baseline é estatisticamente significativo ($p < 0.001$), enquanto o refinamento adicional (Modelo 3) não trouxe melhorias que justificassem a mudança, indicando que o ponto ótimo de parametrização foi atingido.

**Interpretação e Relevância das Variáveis**
Diferente de abordagens anteriores, o XGBoost concentrou sua capacidade preditiva em atributos sintéticos de alta densidade de informação:
* **risk_fh_p (42.29%) e global_prob_risk (36.56%):** Dominam amplamente a estrutura de decisão, o que comprova que a transformação de dados brutos em probabilidades de risco (Engenharia de Dados) foi o principal motor de ganho do modelo.
* **continuous_risk_score (6.11%):** Aparece como o terceiro pilar, reforçando a consistência dessa métrica em diferentes arquiteturas de boosting.
* **Fatores de Estilo de Vida:** Atividade física (`physical_activity_minutes`) e o respectivo score contínuo (`risk_paw_cont`) somam cerca de 4%, mantendo-se como os fatores comportamentais mais relevantes para o algoritmo.

**Conclusões Práticas**
* **Vantagem do Tuning:** O uso de busca sistemática (Modelo 2) provou-se superior ao Baseline com altíssima confiança estatística ($p = 0.0007$), tornando-se a configuração recomendada para implementação.
* **Estabilidade do Threshold:** O threshold otimizado situou-se em **0.51**, muito próximo ao padrão de 0.5. Isso indica um modelo bem equilibrado nativamente, onde o ajuste fino de probabilidade tem impacto marginal na acurácia global, mas estabiliza a matriz de confusão.
