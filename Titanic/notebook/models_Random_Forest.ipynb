## 2. Dataload & Preprocessamento

```python
BASE = Path.cwd().parent   
# =====================================================
# âš™ï¸ 0. carregamento dos preprocessador 
# =====================================================
PP = preprocessador_titanic()

preprocessador
# # =====================================================
# # ğŸ“ 1. Leitura dos dados & SeparaÃ§Ã£o das bases
# # =====================================================

DATA_DIR = BASE / "data" / "processed"
X_train = pd.read_csv(DATA_DIR / "X_train_final.csv").reset_index(drop=True)
X_test  = pd.read_csv(DATA_DIR / "X_test_final.csv")
y_train = pd.read_csv(DATA_DIR / "y_train_final.csv").values.ravel()
y_test  = pd.read_csv(DATA_DIR / "y_test_final.csv")


# =====================================================
#  ğŸ¤– 3.DefiniÃ§Ã£o dos Modelos
# =====================================================
model_RF = RandomForestClassifier(n_estimators=100, max_depth=8, random_state=42)
pipe_RF       = pipe_models(model_RF,PP)
```

## 3.Treinamento

```python
cv_scores = cross_val_score(pipe_RF , X_train, y_train, cv=5)
print(f"Average CV Accuracy: {np.mean(cv_scores)*100:.2f}%") 

# 2. Testa a performance 
pipe_RF.fit(X_train, y_train)
y_pred=pipe_RF.predict(X_test)
print(f"{'='*70}")
print(f"ğŸ¯ Random Forest PadrÃ£o")
print(f"{'='*70}")
print(f"ğŸ“Š **AcurÃ¡cia no Teste**: {accuracy_score(y_test, y_pred):.4f}")
print(f"\nğŸ“‹ **RelatÃ³rio de ClassificaÃ§Ã£o**:")
print(classification_report(y_test, y_pred))
cm=confusion_matrix(y_test, y_pred)
print(f"ğŸ¯ **Matriz de ConfusÃ£o**:")
print(f"               Previsto 0   Previsto 1")
print(f"Real 0         {cm[0,0]:<11} {cm[0,1]:<11}")
print(f"Real 1         {cm[1,0]:<11} {cm[1,1]:<11}")
print(f"{'â”€'*70}")
```
